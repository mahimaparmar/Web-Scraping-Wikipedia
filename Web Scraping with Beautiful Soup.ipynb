```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

print(soup.prettify())

```

```python
table = soup.find('table', class_ = "wikitable sortable sticky-header-multi sort-under")

print(table)
```

```python
world_titles = table.find_all('th', scope = "col")
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
```

```python
world_table_titles = world_table_titles[:-2]
```

```python
world_table_titles
```

```python
for ind in range(0,len(world_table_titles)):
    if '[' in world_table_titles[ind] and ']' in world_table_titles[ind]:
        world_table_titles[ind] = world_table_titles[ind][:world_table_titles[ind].index('[')]
```

```python
world_table_titles
```

```python
df = pd.DataFrame(columns = world_table_titles)
```

```python
df
```

```python
table_rows = table.find_all('tr')[2:]
```

```python
table_rows[0]
```

```python
for row in table_rows:
    row_data = row.find('th')
    row_data.extend(row.find_all('td'))
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
